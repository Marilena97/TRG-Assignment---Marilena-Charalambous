```python
#A function to Find Count of Null, None, NaN of All DataFrame Columns
def CountNull(dataframe):    
    #https://sparkbyexamples.com/pyspark/pyspark-find-count-of-null-none-nan-values/
    from pyspark.sql.functions import col,isnan, when, count
    return dataframe.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in dataframe.columns]
       ).show()
```

```python
#print basic data exploration
def BasicDataExploration(dataframe,name):
    #dataframe name
    print('\033[1m' + name + '\n')
    
    #dataframe size
    print('\033[0m'+'The dataframe contain {} rows and {} columns.'.format(dataframe.count(), len(dataframe.columns)))
    
    ##check the schema
    print('Dataframe Schema:')
    dataframe.printSchema()
    
    #nan
    print('Count of Null, None, NaN of All DataFrame Columns:')
    CountNull(dataframe)
    
    #data describe
    #print('Statistical data of All DataFrame Columns:')
    #dataframe.describe().show()
    
    dataframe.show(10)
```

```python
# Create a file with the district
def exctractdistrict(filename):
    #we exctract the street dataframe
    dfStreet = spark.read.option('header','true').csv(filename,inferSchema=True)
    
    # drop na values on Crime ID
    dfStreet = dfStreet.na.drop(subset='Crime ID')
    
    #base the location we find the outcome dataframe
    Outcomefilename = filename.replace('street','outcomes')

    #if outcomes file exists
    if exists(Outcomefilename):    
        
        #we exctract the outcomes dataframe
        dfOutcome = spark.read.option('header','true').csv(Outcomefilename,inferSchema=True)
        DataOutcome = dfOutcome[[
                        'Crime ID',
                        'Outcome type'
                              ]]
        DataStreet = dfStreet[[
                       'Crime ID',
                       'Latitude',
                       'Longitude',
                       'Crime type',
                       'Last outcome category'
                      ]]
        
        #BasicDataExploration(DataStreet,filename)
        
        df = DataStreet.join(DataOutcome, on=['Crime ID'], how='left')
        
        #where 'Outcome type' is nan fill with "Last outcome category"
        #https://stackoverflow.com/questions/51952584/in-pyspark-is-it-possible-to-fillna-with-another-column
        df = df.withColumn('Outcome type', coalesce('Outcome type', 'Last outcome category'))
        
        df = df.withColumnRenamed('Outcome type','lastOutcome')
    else:
        df  = dfStreet[[
                       'Crime ID',
                       'Latitude',
                       'Longitude',
                       'Crime type',
                       'Last outcome category'
                      ]]
        df = df.withColumnRenamed('Last outcome category','lastOutcome')
        
    
    #get district name 
    #choose last file 
    CsvFile = basename(filename)
    #remove numbers
    CsvWithoutDate = re.sub(r'\d','',CsvFile)
    #remove the ending and -
    distName = CsvWithoutDate.replace('-street.csv','').replace('-',' ').strip()
    df=df.select('*' ,lit(distName).alias("districtName"))
    
    finalset = df.select(
        "Crime ID",
        'districtName',
        'Latitude',
        'Longitude',
        'Crime type',
        'lastOutcome')
    
    #BasicDataExploration(finalset,CsvFile.replace('-street.csv',''))
    
    return finalset
    
```

```python
# explicit function
#https://www.geeksforgeeks.org/merge-two-dataframes-in-pyspark/
def unionAll(dfs):
    return functools.reduce(lambda df1, df2: df1.union(df2.select(df1.columns)), dfs)
```

```python
def transform(path):
    li = []
    #https://www.codegrepper.com/code-examples/python/read+all+csv+files+in+folder+python+pandas
    all_files= glob.glob(path + "/*.csv")
    streetfiles = [ string for string in all_files if  "street" in string]
    
    for i in streetfiles:
        
        df = exctractdistrict(i)
            
        # merge data
        if li == []:
            li=df
        else:
            li=unionAll([li, df])

    BasicDataExploration(li,basename(path))

    return li
```

```python
%%time
#https://www.adamsmith.haus/python/answers/how-to-list-immediate-subdirectories-in-python
folders = os.listdir(datapath)

lis = []
for i in folders:    
    ReadCsv = transform(datapath+ "\\"+i)
    
    if ReadCsv != []:
        
        print(ReadCsv)
        print(type(ReadCsv))
        DataCsv = ReadCsv.toPandas()
    
        # merge data
        # i use pandas because spark did not run this step
        lis.append(DataCsv)

# i used pandas to connect the final dataframe to show my different set of skills
results = pd.concat(lis, axis=0, ignore_index=True)
  
```
